```python
import os
import numpy as np
import pandas as pd
import json
from collections import defaultdict
%matplotlib notebook

from data.irhythm.dataset_tools.db_constants import ECG_SAMP_RATE, ECG_EXT, EPI_EXT, BEA_EXT
from data.irhythm.dataset_tools.signal_plotting import pltWithButton
```

###Read in data:

```python
# root = '/path/to/record/dir'  # <-- fill in with path to record directory 
rec_id = ''  # <-- fill in the desired record id
qa = '_post'
```

```python
ecg_file = os.path.join(root, rec_id+ECG_EXT)
epi_file = os.path.join(root, rec_id+qa+EPI_EXT)
bea_file = os.path.join(root, rec_id+qa+BEA_EXT)
```

```python
# read in data
# raw ecg
ecg = np.fromfile(ecg_file, np.int16)
# R-peaks locations
R_peaks = np.fromfile(bea_file, np.int64)

# rhythm labels
with open(epi_file, 'r') as f:
    ee = json.load(f)

rhy_labels = pd.DataFrame(ee['episodes'])
```

### Interactive plot: 

```python
rhy_dict = defaultdict(list)
for idx, row in rhy_labels.iterrows():
    rhy_dict[row['rhythm_code']].append((row['onset'], row['offset']))

fig, ax, bforward, bbackward, rhy_prev_handles, rhy_next_handles, \
check, slider = pltWithButton(ecg, ECG_SAMP_RATE, 0, 16,
                              {'R': R_peaks}, rhy_dict)
```
